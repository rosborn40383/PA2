```python
x=int(input("Enter Subjects age (0 to quit): "))
while x>0:
    if x<18:
        print("Minor")
    elif x>=18 and x<=64:
        print("Adult")
    elif x>=65:
        print("Senior citizen")
    x=int(input("Enter Subjects age (0 to quit): "))
```

```python
age=[65,78,34,56,12,9,43,67]
for age in age:
    if age<18:
        print(age, "Minor")
    elif age>=18 and age<=64:
        print(age, "Adult")
    elif age>=65:
        print(age, "Senior")
```
